```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\aleks\\Desktop\\courseML\\05-Seaborn\\dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize = (12,4), dpi = 200)
sns.scatterplot(x = 'salary', y= 'sales',data = df, style = 'level of education',hue ='level of education' )
```

```python

sns.rugplot(x = 'salary', data = df, height = 0.5)
```

```python
sns.set(style = 'ticks')
sns.displot(data = df, x = 'salary',bins = 50, color = 'red', edgecolor = 'blue', kde = True, rug = True)
```

```python
sns.histplot(data = df, x = 'salary')
```

```python
plt.figure(figsize =(10,4))
sns.countplot(data = df, x = 'level of education', hue = 'division')
```
