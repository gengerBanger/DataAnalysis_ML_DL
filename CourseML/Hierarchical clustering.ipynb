```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cluster_mpg.csv')
```

```python
df
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_dummies = pd.get_dummies(df.drop('name', axis = 1))
```

```python
df_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler # minmaxscaler подходит 
                                               #лучше для алгоритма иерархической кластеризации
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_dummies) # получаем двумерный массив
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df_dummies.columns)#преобразуем в дата фрейм
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(scaled_df, cmap = 'viridis')
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df, cmap = 'viridis',row_cluster=False) 
```

```python
sns.heatmap(scaled_df.corr())
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df, cmap = 'viridis',col_cluster=False) 
```

```python
from sklearn.cluster import AgglomerativeClustering # смотри ноушен
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(15,8), dpi = 200)
sns.scatterplot(data = df, x ='mpg', y = 'weight', hue = cluster_labels)
```

```python
plt.figure(figsize=(15,8), dpi = 200)
sns.scatterplot(data = df, x ='mpg', y = 'horsepower', hue = cluster_labels, palette = 'viridis')
```

```python
# считаем максимальное расстояние между точками для параметра  distance_threshhold
np.sqrt(len(scaled_df.columns))
```

```python
model = AgglomerativeClustering(n_clusters = None, distance_threshold=2)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix # получаем списки иерархических кластеров
#67 и 161 - номера кластеров, 1.41421356 - расстояние между кластерами, 2 - количество
# точек в получившемся кластере 
```

```python
plt.figure(figsize = (20,10), dpi = 250)
dendro = dendrogram(linkage_matrix, truncate_mode = 'level', p = 4) # усечение по уровням иерархии
```

```python
plt.figure(figsize = (20,10), dpi = 250)
dendro = dendrogram(linkage_matrix, truncate_mode = 'lastp', p = 11) # усечение по количеству кластеров
```
