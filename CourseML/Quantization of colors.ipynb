```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg # .jpg, .png -> Массив numpy
```

```python
image_as_array = mpimg.imread('../DATA/yogurt.jpg')
```

```python
image_as_array.shape # H, W, C
```

```python
image_as_array
```

```python
plt.Figure(figsize = (12,6), dpi = 300)
plt.imshow(image_as_array)
```

```python
(h, w, c) = image_as_array.shape
```

```python
h
```

```python
w
```

```python
c
```

```python
image_as_array2d = image_as_array.reshape(h*w,c)
```

```python
len(image_as_array2d.shape)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=10)
labels = model.fit_predict(image_as_array2d)
```

```python
labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
rgb_codes[labels]
```

```python
image_as_array2d
```

```python
quntized_image = np.reshape(rgb_codes[labels], (h,w,c))
```

```python
quntized_image
```

```python
plt.imshow(quntized_image)
```
