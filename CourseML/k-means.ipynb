```python
X = pd.get_dummies(df)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
help(KMeans)
```

```python
model = KMeans(n_clusters = 2)
```

```python
cluster_labels = model.fit_predict(scaled_X)
```

```python
cluster_labels
```

```python
X['Cluster'] = cluster_labels
```

```python
X
```

```python
plt.figure(figsize=(12,6), dpi =200)
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind = 'bar')
```

```python
ssd = []
for k in range(2,10):
    model = KMeans(n_clusters = k)
    model.fit(scaled_X)
    
    ssd.append(model.inertia_) #ÑÑƒÐ¼Ð¼Ð° ÐºÐ²Ð°Ð´Ñ€Ð°Ñ‚Ð¾Ð² Ñ€Ð°ÑÑÑ‚Ð¾ÑÐ½Ð¸Ð¹
```

```python
ssd
```

```python
plt.plot(range(2,10),ssd,'o--')
```

```python
ssd_series = pd.Series(ssd)
```

```python
ssd_series.diff()
```

```python
pd.DataFrame(data = zip(range(2,10),ssd_series), columns = ['K','SSD diff'])
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouettes = []
for k in range(2,10):
    model = KMeans(n_clusters = k)
    model.fit(scaled_X)
    
    silhouettes.append(silhouette_score(scaled_X, model.labels_))
```
