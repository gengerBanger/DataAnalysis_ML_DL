# Задача 3
Ночью произошло ДТП с участием такси. В городе работают две компании такси — «Синие»(1) и «Зелёные(0)». «Зелёным» принадлежит 85% такси, «Синим» 15%. Свидетель аварии утверждает, что такси принадлежало «Синим». Следственный эксперимент показал, что ночью свидетель верно определяет цвет такси в 80%. Какова вероятность того, что такси действительно принадлежало «Синим»?

```python
def observer (taxi):
  if np.random.binomial(1,.8):
    return taxi
  return abs(taxi - 1)
```

```python
observer(1)
```

```python
n = 10000
result = []

for _ in range(n):
  taxi = np.random.binomial(1,.15)
  answer = observer(taxi)
  result.append((taxi, answer))
```

```python
 pd.DataFrame(result, columns=['taxi','answer'])
```

```python
 pd.DataFrame(result, columns=['taxi','answer']).groupby('answer').agg(['mean'])
```

# Задача 4

Русская рулетка. Есть револьвер с 2 вставленными подряд патронами. Первый крутит барабан и стреляет и остается жив. Твоя очередь: тебе предлагает покрутить барабан перед выстрелом. Согласишься или сразу выстрелишь и почему?

```python
chamber = [0,0,0,1,0,1]
```

```python
def one_spin(chamber):
  n = len(chamber)
  new_chamber = [0,0,0,0,0,0]
  for i in range(n):
    if i != n-1:
      new_chamber[i+1] = chamber[i]
    else:
      new_chamber[0] = chamber[i]

  return new_chamber
```

```python
one_spin(chamber)
```

```python
def n_spin(chamber):
  n = np.random.randint(1,15)
  for i in range(n):
    chamber = one_spin(chamber)
  return chamber
```

```python
n_spin(chamber)
```

```python
n = 100000
result = []
chamber = [1,1,0,0,0,0]
for _ in range(n):
  spinned_chamber = n_spin(chamber)
  spinned_chamber = one_spin(spinned_chamber)
  player_1 = spinned_chamber[0]
  if player_1:
    continue
  spinned_chamber = n_spin(spinned_chamber)
  spinned_chamber = one_spin(spinned_chamber)
  player_2 = spinned_chamber[0]
  result.append([player_2])

np.mean(result) # вероятность словить маслину
```
