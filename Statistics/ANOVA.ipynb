2. Автодилер проводит сравнение трех различных брендов автомобилей по их топливной эффективности. У него есть данные о расходе топлива на 100 км для каждого бренда. Нужно определить, есть ли статистически значимые различия между брендами автомобилей.

3. Ботаники проводят эксперимент с тремя разными дозами удобрений для растений. Они измеряют рост растений после использования каждой из доз. Цель - определить, оказывают ли эти различные дозы статистически значимое влияние на рост растений.


image.png

```python
import scipy.stats as stats

# Группы данных
group1 = [4,5,3,4,5]
group2 = [6,7,8,6,7]
group3 = [5,4,6,5,4]

# Выполнение однофакторного ANOVA
f_stat, p_value = stats.f_oneway(group1, group2, group3)

# Печать результатов
print("F-статистика:", f_stat)
print("p-value:", p_value)
```

```python
import scipy.stats as stats

# Группы данных
group1 = [7.8, 8.2, 7.9, 8.0, 8.1]
group2 = [8.5, 8.6, 8.4, 8.5, 8.7]
group3 = [8.0, 7.9, 8.1, 8.2, 7.8]

# Выполнение однофакторного ANOVA
f_stat, p_value = stats.f_oneway(group1, group2, group3)

# Печать результатов
print("F-статистика:", f_stat)
print("p-value:", p_value)
```
