```python
def percent_of_Nan(dataFrame):
    result = 100 * dataFrame.isnull().sum() / len(dataFrame)
    result = result[result>0].sort_values()
    plt.figure(figsize = (12,8), dpi = 250)
    sns.barplot(x = result.index, y = result)
    print(result)
    plt.xticks(rotation = 90)
```

```python
X = df.drop('SalePrice', axis = 1)
y = df['SalePrice']
X
```

```python
dum_X = pd.get_dummies(data = X, drop_first=True)
```

```python
dum_X
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(dum_X, y, test_size=0.10, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
elastic_model = ElasticNet(max_iter=1000000)
```

```python
scaled_X_train.shape
```

```python
param_grid = {'alpha':[9],
             'l1_ratio':[0.9473684210526315]}
grid_search = GridSearchCV(estimator=elastic_model,
                          param_grid=param_grid,
                          scoring = 'neg_mean_squared_error', 
                          cv = 10,
                          verbose=3)
```

```python
grid_search.fit(scaled_X_train, y_train)
```

```python
grid_search.best_params_
```

```python
scaled_X_test
```

```python
pred = grid_search.predict(scaled_X_test)
```

```python
y_test
```

```python
pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
np.sqrt(mean_squared_error(y_test,pred))
```

```python
mean_absolute_error(y_test, pred)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gr_boost = GradientBoostingRegressor(n_estimators=195)
```

```python
gr_boost.fit(scaled_X_train, y_train)
```

```python
boost_pred = gr_boost.predict(scaled_X_test)
```

```python
mean_absolute_error(y_test, boost_pred)
```

```python
np.sqrt(mean_squared_error(y_test,boost_pred))
```
