```python
#!c1.32
total_df = total_df[total_df['Law Section'] < 1200]
```

```python
#!c1.32
total_df = total_df[(total_df['Law Section'] != 1040.0) & (total_df['Law Section'] != 40.0)]
```

```python
#!c1.32
total_df['Issuer Precinct'].value_counts()
```

```python
#!c1.32
X = total_df[['Registration State','Plate Type','Violation Code','Issuing Agency',
              'Street Name','Sub Division','Street Code1','Street Code2','Street Code3']]
y = total_df['Law Section']
```

```python
#!c1.32
X['Feet From Curb'].unique()
```

```python
#!c1.32
from sklearn.model_selection import train_test_split
```

```python
#!c1.32
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.45, random_state=42)
```

```python
#!c1.32
```

```python
#!c1.32
from sklearn.ensemble import GradientBoostingClassifier
from catboost import CatBoostClassifier
```

```python
#!c1.32
total_df.head()
```

```python
#!c1.32
model = CatBoostClassifier(cat_features = ['Registration State','Plate Type','Violation Code','Issuing Agency',
                                          'Street Name','Sub Division'], iterations=150, depth=6, learning_rate=0.3,)
```

```python
#!c1.32
model.fit(X_train, y_train)
```

```python
#!c1.32
pred = model.predict(X_test)
```

```python
#!c1.32
from sklearn.metrics import classification_report
```

```python
#!c1.32
print(classification_report(y_test, pred))
```

```python
#!c1.32
import time
model = CatBoostClassifier(cat_features = ['Registration State','Plate Type','Violation Code','Issuing Agency',
                                          'Street Name','Sub Division'], iterations=100, depth=6, learning_rate=0.3,)
```

```python
#!c1.32
time_count = []
for i in [.1, .25, .5, .75, 1]:
    length = int(X_train.shape[0]*i)
    start_time = time.time()
    model.fit(X_train[:length], y_train[:length], verbose=25)
    train_time = time.time() - start_time
    time_count.append(train_time)
```

```python
#!c1.32
time_count
```

```python
#!c1.32
plt.figure(figsize = (5,3), dpi = 200)
sns.barplot(x = ['10%', '25%', '50%', '75%', '100%'], y = time_count)
plt.xticks(rotation = 60)
plt.ylabel(' , ')
plt.xlabel('   , %')
```

```python
#!c1.32
plt.figure(figsize = (5,3), dpi = 200)
plt.plot([10, 25, 50, 75, 100],time_count, marker = 'o')
plt.xticks(rotation = 60)
plt.ylabel(' , ')
plt.xlabel('   , %')
```
